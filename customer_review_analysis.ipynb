#Customer Review Sentiment Analysis using Machine Learning

This project analyzes customer reviews using machine learning techniques to extract sentiment and identify patterns that can provide business insights.

## Research Question

How can AI be used to analyze customer reviews and extract useful insights for decision-making?

```python
!pip install datasets scikit-learn

import pandas as pd
from datasets import load_dataset
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
#Load Dataset
dataset = load_dataset("amazon_polarity")

data = dataset["train"].shuffle(seed=42).select(range(2000))

texts = data["content"]
labels = data["label"]

df = pd.DataFrame({"review": texts, "sentiment": labels})
df.head()
```

## Dataset

We use a subset of the Amazon Reviews dataset. Each review is labeled as positive or negative, allowing us to train a sentiment analysis model.

```python
#Model Evaluation (Train-Test Split)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df["review"], df["sentiment"], test_size=0.2, random_state=42
)

vectorizer = TfidfVectorizer(max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = LogisticRegression()
model.fit(X_train_vec, y_train)

preds = model.predict(X_test_vec)

from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(y_test, preds))
```

To properly evaluate the model, we split the dataset into training and test sets.
This prevents the model from simply memorizing the data and provides a more realistic performance estimate.

```python
# Show sample predictions
sample_df = pd.DataFrame({
    "review": X_test[:5],
    "true_label": y_test[:5],
    "predicted": preds[:5]
})

sample_df
```

## Results

The Logistic Regression model achieved an accuracy of approximately 0.77 on the test set.

Sample predictions show that the model can correctly classify both positive and negative reviews, although some errors remain.


## Interpretation

The model demonstrates that even simple machine learning techniques can effectively capture sentiment patterns in text data.

However, performance is limited compared to more advanced deep learning approaches, which could better capture context and nuance.
